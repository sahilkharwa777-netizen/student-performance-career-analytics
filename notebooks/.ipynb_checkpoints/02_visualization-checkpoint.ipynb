```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your actual file
df = pd.read_excel("../data/Power BI Updated Data analyst Data.xlsx")

# Convert to numeric
cols = ["CGPA", "Expected salary (Lac)", "Experience with python (Months)", "Family Income"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Clean data
df = df.dropna(subset=["CGPA", "Expected salary (Lac)"])
```

```python
print(df.shape)
```

```python
plt.figure()
sns.histplot(df["CGPA"], bins=10)
plt.title("Distribution of CGPA")
plt.xlabel("CGPA")
plt.ylabel("Number of Students")
plt.show()
```

Most students fall within a certain CGPA range, indicating overall academic performance trends.

```python
plt.figure()
sns.scatterplot(x=df["CGPA"], y=df["Expected salary (Lac)"])
plt.title("CGPA vs Expected Salary")
plt.xlabel("CGPA")
plt.ylabel("Expected Salary (Lac)")
plt.show()
```

Higher CGPA students tend to expect higher salaries, showing a positive relationship.

```python
plt.figure(figsize=(8,5))
df["City"].value_counts().plot(kind='bar')
plt.title("Number of Students by City")
plt.xlabel("City")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Some cities contribute more students, indicating regional participation differences.

```python
plt.figure()
sns.boxplot(x=df["Leadership- skills"], y=df["CGPA"])
plt.title("Leadership Skills vs CGPA")
plt.show()
```

Students with leadership skills may show slightly higher academic performance.

```python
plt.figure()
sns.scatterplot(x=df["Experience with python (Months)"], y=df["Expected salary (Lac)"])
plt.title("Python Experience vs Expected Salary")
plt.xlabel("Python Experience (Months)")
plt.ylabel("Expected Salary (Lac)")
plt.show()
```

More experience in Python tends to correlate with higher salary expectations.

```python
top_colleges = df.groupby("College Name")["CGPA"].mean().sort_values(ascending=False).head(5)

plt.figure()
top_colleges.plot(kind='bar')
plt.title("Top 5 Colleges by Average CGPA")
plt.ylabel("CGPA")
plt.xticks(rotation=45)
plt.show()
```

Certain colleges consistently produce higher-performing students.
